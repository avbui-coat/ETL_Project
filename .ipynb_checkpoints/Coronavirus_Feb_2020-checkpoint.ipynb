```python
import pandas as pd
import numpy as np
```

```python
main_df = pd.read_csv("2019_nCoV_data.csv")
confirmed_df = pd.read_csv("time_series_covid_19_confirmed.csv")
recovered_df = pd.read_csv("time_series_covid_19_recovered.csv")
death_df = pd.read_csv("time_series_covid_19_deaths.csv")
```

```python
main_df
```

```python
split_data = main_df["Date"].str.split(" ",expand=True)
# data = split_data.to_list()
# names = ["Date", "Time"]
# main_cleaned_df = pd.DataFrame(data, columns=names)
split_data.head()
```

```python
main_df['Date Only']=split_data[0]
main_df['Time']=split_data[1]
main_df.drop(columns =["Date"], inplace = True) 
main_df.head()
```

```python
main_df['Province/State'].unique()
```

```python
main_df['Province/State'].value_counts()
```

```python
main_df['Country'].value_counts()
```

```python
main_df['Province/State'] = main_df['Province/State'].replace(
    {'Cruise Ship': 'Diamond Princess cruise ship',
     'Chicago': 'Chicago, IL',
    'California':'Sacramento, CA',
    'San Diego County, CA':'San Diego, CA',
    'Washington': 'Seattle, WA',
    'Illinois': 'Chicago, IL',
     'Ontario': 'Toronto, ON',
     'Victoria':'Victoria, BC',
     'Arizona': 'Tempe, AZ'
    })
```

```python
main_df['Province/State'].unique()
```

```python
main_df.to_csv('fairly_cleaned_main.csv')
```
